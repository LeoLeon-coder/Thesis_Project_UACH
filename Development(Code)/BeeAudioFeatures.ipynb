```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# Load audio
sr, audio = wavfile.read(r"C:\Users\leona\Documents\Thesis_Project_UACH\Temp\Dataset\BeesAnna\sound_files\2022-06-06--14-24-42_2__segment2.wav")
```

```python
audio.shape
```

```python
audio.dtype
```

```python
# Normalize if int16
if audio.dtype == np.int16:
    audio = audio.astype(np.float32) / np.iinfo(np.int16).max
```

```python
# Create time axis
duration_sec = len(audio) / sr
time_axis = np.linspace(0, duration_sec, num=len(audio))
```

```python
# Plot waveform
plt.figure(figsize=(12, 4))
plt.plot(time_axis, audio)
plt.title(f"Waveform (Sample Rate: {sr} Hz)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

The file is a mono, 16-bit PCM WAV file sampled at 22050 Hz.
